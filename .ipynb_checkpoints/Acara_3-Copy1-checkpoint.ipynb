Import LIbraries dan Cek Direktori

```python
import os
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
import rasterio
from rasterio.transform import from_origin
import numpy as np
from rasterstats import zonal_stats
```

```python
# Cek Direktori
print("Direktori kerja saat ini:", os.getcwd())
```

Read Data NetCDF

```python
# Path file NetCDF
file_nc_path = 'D:/Geodatabase/03_Ekstraksi_Data/01_Raw_Data/gpcp_v01r03_daily_d20230101_c20230411.nc'

# Membuka file NetCDF dengan Xarray
try:
    data = xr.open_dataset(file_nc_path, engine='netcdf4')
except ValueError as e:
    print(f"Error saat membuka file: {e}")
    print("Pastikan library netCDF4 sudah terinstal.")
    exit ()
    
# Menampilkan informasi umum dataset
print("=== Informasi Dataset ===")
print(data)

# Menampilkan variabel yang tersedia
print("\n=== Variabel ===")
for var in data.data_vars:
    print(f"Nama: {var}, Dimensi: {data[var].dims}, Atribut: {data[var].attrs}")
```

NetCDF diubah ke GeoTiff

```python
# Direktori input dan output
input_nc_folder = r"D:\Geodatabase\03_Ekstraksi_Data\01_Raw_Data"  # Ganti dengan lokasi folder file NC
output_tiff_folder = r"D:\Geodatabase\03_Ekstraksi_Data\02_GeoTiff_Data"  # Ganti dengan lokasi folder output

# Membuat folder output jika belum ada
os.makedirs(output_tiff_folder, exist_ok=True)
```

```python
# Loop melalui semua file di folder
for file in os.listdir(input_nc_folder):
    if file.endswith(".nc"):
        file_path = os.path.join(input_nc_folder, file)

        # Ekstrak tanggal dari nama file
        date_str = file.split("_")[3][1:]  # Mengambil "YYYYMMDD" dari nama file
        output_file = os.path.join(output_tiff_folder, f"precip_{date_str}.tif")

        # Buka file NC
        with xr.open_dataset(file_path) as ds:
            # Pastikan variabel yang digunakan adalah "precip"
            if "precip" in ds.variables:
                precip = ds["precip"]

                # Pilih data untuk waktu pertama (index 0)
                precip = precip.isel(time=0).values  # Ambil data pertama pada dimensi waktu

                # Koordinat latitude dan longitude
                lats = ds["latitude"].values
                lons = ds["longitude"].values

                # Normalisasi longitude dari [0, 359] menjadi [-180, 180]
                lons = np.where(lons > 180, lons - 360, lons)
                sorted_indices = np.argsort(lons)  # Mendapatkan indeks untuk mengurutkan longitude
                lons = lons[sorted_indices]  # Mengurutkan longitude

                # Menyesuaikan grid data berdasarkan longitude yang telah diurutkan
                precip = precip[:, sorted_indices]

                # **Set nilai di bawah 0 menjadi 0**
                precip = np.maximum(precip, 0)

                # Georeferensi: asumsi lat/lon dalam derajat
                res_lat = abs(lats[1] - lats[0])
                res_lon = abs(lons[1] - lons[0])
                transform = from_origin(lons.min(), lats.max(), res_lon, res_lat)

                # Validasi bentuk data
                if precip.ndim != 2:
                    raise ValueError(f"Bentuk data tidak valid untuk GeoTIFF: {precip.shape}")

                # Tulis ke GeoTIFF
                with rasterio.open(
                    output_file,
                    "w",
                    driver="GTiff",
                    height=precip.shape[0],
                    width=precip.shape[1],
                    count=1,
                    dtype=np.float32,
                    crs="EPSG:4326",  # CRS WGS84
                    transform=transform,
                ) as dst:
                    dst.write(precip, 1)

        print(f"Berhasil mengonversi: {file} -> {output_file}")
```

Buat Data Presipitasi Bulanan

```python
# Direktori input dan output
input_daily_folder = r"D:\Geodatabase\03_Ekstraksi_Data\02_GeoTiff_Data"  # Ganti dengan folder input
output_monthly_folder = r"D:\Geodatabase\03_Ekstraksi_Data\03_Monthly_Data"  # Ganti dengan folder output

# Membuat folder output jika belum ada
os.makedirs(output_monthly_folder, exist_ok=True)
```

```python
# Mengambil daftar file dan mengelompokkan berdasarkan bulan
files = sorted([f for f in os.listdir(input_daily_folder) if f.endswith('.tif')])
monthly_files = {}

for file in files:
    year_month = file[7:13]  # Ekstrak YYYYMM dari nama file
    if year_month not in monthly_files:
        monthly_files[year_month] = []
    monthly_files[year_month].append(os.path.join(input_daily_folder, file))
```

```python
# Penjumlahan file raster per bulan
for year_month, file_list in monthly_files.items():
    print(f"\nProcessing month: {year_month}")
    monthly_sum = None
    meta = None

    for file in file_list:
        with rasterio.open(file) as src:
            data = src.read(1)  # Baca band pertama
            if monthly_sum is None:
                # Inisialisasi array untuk penjumlahan
                monthly_sum = np.zeros_like(data, dtype=np.float32)
                meta = src.meta  # Ambil metadata dari file pertama
            monthly_sum += data

    # Update metadata untuk file output
    meta.update(dtype=rasterio.float32, count=1)  # Tidak menggunakan kompresi

    # Simpan hasil ke file baru tanpa kompresi
    output_monthly_file = os.path.join(output_monthly_folder, f"precip_{year_month}.tif")
    with rasterio.open(output_monthly_file, 'w', **meta) as dst:
        dst.write(monthly_sum, 1)
    
    # Tampilkan keterangan file berhasil disimpan
    print(f"File precip_{year_month}.tif telah disimpan di {output_monthly_folder}")
```

Zonal Statistic

```python
# Define folder containing raster files and the shapefile
raster_folder = r"D:\Geodatabase\03_Ekstraksi_Data\03_Monthly_Data"
shapefile_path = r"D:\Geodatabase\03_Ekstraksi_Data\04_Zonal_Statistik\01_Admin_Data\Jawa_Provinsi.shp"
```

```python
# Load shapefile
gdf = gpd.read_file(shapefile_path)

# Menampilkan kolom yang ada di shapefile
print("Kolom yang tersedia dalam shapefile:")
print(gdf.columns)

# Jika ingin menampilkan beberapa data contoh
print("\nBeberapa baris data:")
print(gdf.head())
```

```python
# Pastikam Geometri Valid
gdf = gdf[gdf.is_valid]
```

```python
# Set unique identifier field
gdf.set_index('Province', inplace=True)
```

```python
# Peroleh List File Raster
raster_files = sorted(f for f in os.listdir(raster_folder) if f.startswith('precip_') and f.endswith('.tif'))

# Print list raster files
print(raster_files)
```

```python
# Iterasi melalui setiap file raster
for raster_file in raster_files:
    try:
        # Ekstraksi informasi tahun dan bulan dari nama file
        year_month = raster_file.split('_')[1].split('.')[0]  # Misalnya: 'yyyymm'
        month = year_month[4:6]  # Ekstraksi 'mm' dari 'yyyymm'
    except IndexError:
        print(f"Error dalam memproses file {raster_file}. Skipping.")
        continue  # Lewati file ini jika format nama tidak sesuai

    raster_path = os.path.join(raster_folder, raster_file)
    
    with rasterio.open(raster_path) as src:
        # Ambil transformasi affine, data array, dan nilai nodata
        affine = src.transform
        array = src.read(1)  # Membaca band pertama
        nodata_value = src.nodata  # Ambil nilai nodata dari metadata
            
        # Pastikan nilai nodata tidak kosong
        if nodata_value is None:
                nodata_value = -999  # Tetapkan default jika tidak ada

        # Menghitung zonal statistics
        stats = zonal_stats(
            gdf,
            array,
            affine=affine,
            nodata=nodata_value,  # Berikan nodata secara eksplisit
            stats='mean',  # Hitung rata-rata nilai
            all_touched=True  # Sertakan semua piksel yang menyentuh geometri
        )

        # Menambahkan hasil ke GeoDataFrame dengan nama kolom sesuai bulan
        gdf[month] = [s['mean'] for s in stats]
```

```python
# Save dalam format Shapefile
output_shapefile_path = r"D:\Geodatabase\03_Ekstraksi_Data\04_Zonal_Statistik\02_Hasil_Zonal_Statistik\01_Presipitasi_Bulanan_2023.shp"
gdf.to_file(output_shapefile_path)

print(f" Hasil zonal statistic tersimpan di {output_shapefile_path}")
```

Visualisasi Grafik Presipitasi Bulanan

```python
def plot_precip_by_id(file_path, output_folder="output_plots", dpi=300):
    """
    Memvisualisasikan data presipitasi berdasarkan ID Provinsi dari file Excel 
    dan menyimpan plot sebagai gambar di folder yang ditentukan.

    Args:
        file_path (str): Path ke file Excel yang berisi data presipitasi.
        output_folder (str, opsional): Folder untuk menyimpan plot yang dihasilkan. Default adalah "output_plots".
        dpi (int, opsional): Resolusi gambar yang disimpan. Default adalah 300.
    """
    # Membaca file Excel ke dalam DataFrame
    df = pd.read_excel(file_path)

    # Memastikan kolom yang dibutuhkan ada dalam DataFrame
    if 'Province' not in df.columns:
        raise ValueError("File input harus mengandung kolom 'Province'.")

    # Iterasi untuk setiap ID Provinsi unik
    for id_data in df['Province'].unique():
        # Menyaring data berdasarkan ID Provinsi
        df_filtered = df[df['Province'] == id_data]

        # Membuat figure untuk plot
        plt.figure(figsize=(12, 5))  # Ukuran disesuaikan untuk keterbacaan yang lebih baik
        plt.plot(df_filtered.columns[1:], df_filtered.iloc[0, 1:], marker='o')
            
        # Menambahkan judul dan label sumbu dengan font tebal
        plt.title(f"Presipitasi Bulanan - {id_data} - 2023", fontweight='bold')
        plt.xlabel("Bulan", fontweight='bold')
        plt.ylabel("Presipitasi (mm)", fontweight='bold')
            
        # Memutar label sumbu x agar vertikal
        plt.xticks(rotation=90)

        # Menambahkan grid pada plot
        plt.grid(True, linestyle='--', alpha=0.7)

        # Membuat folder output jika belum ada
        os.makedirs(output_folder, exist_ok=True)

        # Menyaring nama ID Provinsi agar bisa digunakan sebagai nama file
        safe_id_data = "".join(c if c.isalnum() or c.isspace() else "_" for c in str(id_data))

        # Menyusun nama file output
        filename = f"{output_folder}/Presipitasi_{safe_id_data}.png"

        # Menyimpan plot sebagai gambar PNG dengan resolusi tinggi
        plt.savefig(filename, dpi=dpi, bbox_inches='tight')  # DPI tinggi dan box yang rapat untuk kualitas
        plt.show()  # Menampilkan plot
        plt.close()  # Menutup figure plot untuk menghindari masalah memori
            
        # Memberikan informasi bahwa file telah disimpan
        print(f"File disimpan: {filename}")
```

```python
# Penerapan
file_path = r"D:\Geodatabase\03_Ekstraksi_Data\04_Zonal_Statistik\02_Hasil_Zonal_Statistik\02_Presipitasi_Bulanan_2023.xlsx"
output_folder =r"D:\Geodatabase\03_Ekstraksi_Data\05_Grafik" # Customize the output folder name (optional)
plot_precip_by_id(file_path, output_folder)
```
